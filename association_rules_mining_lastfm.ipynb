Objective:                                                              
1. To build a system that recommends new music to users in this community from the available information by calculating support, confidence, and lift.                                                 
2. To look for artists (or groups
of artists) who have support larger than 0.01 (1%) and who give confidence to
another artist that is larger than 0.50 (50%) and also lift that is larger than 5            

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
pd.options.display.max_columns = 25
pd.options.display.max_rows = None
```

```python
lastfm = pd.read_csv("https://www.biz.uiowa.edu/faculty/jledolter/DataMining/lastfm.csv")
lastfm_copy = lastfm.copy()
```

```python
lastfm.head(10)
```

```python
#to check if null are there
lastfm.isnull().sum()
```

```python
#to get no: of rows and columns
lastfm.shape
```

```python
#to get the no:of duplicated rows
lastfm.duplicated().sum()
```

```python
#deleted the duplicated rows
lastfm = lastfm.drop_duplicates()
```

```python
#check the shape of dataset after deletion of duplicate rows
lastfm.shape
```

```python
#lastfm['cnt'] = 1
```

```python
#using group by to group the artist column  with respect to user .Using count to get the count as 1 and make it horizontal
lastfm_1 = lastfm.groupby(['user','artist'])['user'].count().unstack()
```

```python
lastfm_1.head(10)
```

```python
#defining a function to convert all the NaN values to 0
def make_boolean(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
#using applymap function to use the created function in the dataset
lastfm_final = lastfm_1.applymap(make_boolean)
```

```python
#matrix is formed and checked the shape
lastfm_final.shape
```

```python
lastfm_final.head(10)
```

```python
#minimum support set is 0.01.Using apriori function to filter rows according to support and item set created according to its frequency of occurance
ap = apriori(lastfm_final, min_support = .01, use_colnames = True)
```

```python
#obtained support for the itemsets
ap.head()
```

```python
#minimum confidence set as 0.5.Here metric is set as confidence
rules = association_rules(ap, metric = "confidence",min_threshold = .5)
```

```python
rules.head()
```

```python
#now filtered with lift greater than 5 and sorting according to decresing oredr of confidence
rules[rules["lift"]>5].sort_values(by='confidence',ascending=False)
#here obtained 6 associations with support > 0.01 and confidence > .50 and lift > 5
```
